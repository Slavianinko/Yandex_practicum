### –ü–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –¥–∞–Ω–Ω—ã—Ö

–ü–æ–¥–≥–æ—Ç–æ–≤–∫—É –¥–∞–Ω–Ω—ã—Ö –ø—Ä–æ–∏–∑–≤–µ–¥—ë–º –≤ —Å–ª–µ–¥—É—é—â–µ–π –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏:

2. –ü–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –≤—ã–±–æ—Ä–æ–∫;
3. –í–µ–∫—Ç–æ—Ä–∏–∑–∞—Ü–∏—è —Ç–µ–∫—Å—Ç–∞.

#### –ü–æ–¥–≥–æ—Ç–æ–≤–∫–∞ —Ç–µ–∫—Å—Ç–∞

–ü—Ä–∏–º–µ–Ω–∏–º –∫ —Ç–µ–∫—Å—Ç–æ–≤—ã–º –¥–∞–Ω–Ω—ã–º —Å–ª–µ–¥—É—é—â–∏–µ –∏–Ω—Å—Ç—Ä—É–º–µ–Ω—Ç—ã –ø–æ–¥–≥–æ—Ç–æ–≤–∫–∏ —Ç–µ–∫—Å—Ç–∞:
1. –û—á–∏—Å—Ç–∫—É –æ—Ç –ª–∏—à–Ω–∏—Ö —Å–∏–º–≤–æ–ª–æ–≤;
2. –õ–µ–º–∞—Ç–∏–∑–∞—Ü–∏—é;
3. –£–¥–∞–ª–µ–Ω–∏–µ —Å—Ç–æ–ø-—Å–ª–æ–≤.

–ü–æ—Å–ª–µ –ø–æ–¥–≥–æ—Ç–æ–≤–∫–∏ —Ç–µ–∫—Å—Ç–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö, –ø—Ä–æ–∞–Ω–∞–ª—Ç–∑–∏—Ä—É–µ–º –∫–∞–∫–∏–µ —Å–ª–æ–≤–∞ —á–∞—â–µ –≤—Å–µ–≥–æ –≤—Å—Ç—Ä–µ—á–∞—é—Ç—Å—è –≤ —Ç–æ–∫—Å–∏—á–Ω—ã—Ö –∏ –≤ –Ω–µ —Ç–æ–∫—Å–∏—á–Ω—ã—Ö –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏—è—Ö

```python
%%time
# –û—á–∏—Å—Ç–∫–∞ —Ç–µ–∫—Å—Ç–∞
wnl = WordNetLemmatizer()
stopwords = set(nltk_stopwords.words('english'))
# –æ—á–∏—Å—Ç–∫–∞ —Å –ª–µ–º–∞—Ç–∏–∑–∞—Ü–∏–µ–π
features = df['text'].apply(lambda x: ' '.join([wnl.lemmatize(re.sub(r'[^a-zA-Z]', ' ', i).lower()) for i in x.split()]))
```

**–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å–ª–æ–≤ –ø–æ –∫–ª–∞—Å—Å–∞–º –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏–µ–≤**

```python
def toxic_cloud(text, label):
    """ –ü—Ä–∏–Ω–∏–º–∞–µ–º —Å–ø–∏—Å–æ–∫ –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏–µ–≤ –∏ –∑–∞–≥–æ–ª–æ–≤–æ–∫
        –†–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ–º —Ç–æ–ø 100 —Å–ª–æ–≤ –∏–∑ —Å–ø–∏—Å–∫–∞
        –í—ã–≤–æ–¥–∏–º –Ω–∞ —ç–∫—Ä–∞–Ω –æ–±–ª–∞–∫–æ —Å–ª–æ–≤
    """
    # –°–æ—Å—Ç–∞–≤–ª—è–µ–º —Å–ø–∏—Å–æ–∫ –≤—Å–µ—Ö —Å–ª–æ–≤ –∏–∑ –ø—Ä–∏–Ω—è—Ç—ã—Ö –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏–µ–≤
    words = []
    for wordlist in text:
        words += wordlist.split()
    # –û—á–∏—â–∞–µ–º –∏—Ö –æ—Ç —Å—Ç–æ–ø-—Å–ª–æ–≤
    clear_words = []    
    for word in words:
        if word not in stopwords:
            clear_words.append(word)
    # –°–æ—Å—Ç–∞–≤–ª—è–µ–º —Å–ø–∏—Å–æ–∫ 100 —Å–∞–º—ã—Ö —á–∞—Å—Ç—ã—Ö —Å–ª–æ–≤
    mostcommon = FreqDist(clear_words).most_common(100)

    # –°–æ–∑–¥–∞—ë–º –æ–±–ª–∞–∫–æ —Å–ª–æ–≤ –∏ –≤—ã–≤–æ–¥–∏–º –Ω–∞ —ç–∫—Ä–∞–Ω
    wordcloud = WordCloud(width=600, height=300, background_color='white').generate(str(mostcommon))
    fig = plt.figure(figsize=(18,7), facecolor='white')
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.axis('off')
    plt.title(label, fontsize=30)
    plt.tight_layout(pad=0)
    plt.show()
```

```python
# To–ø 100 —Å–ª–æ–≤ –≤ —Ç–æ–∫—Å–∏—á–Ω—ã—Ö –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏—è—Ö
toxic_cloud(features[df['toxic'] == 1],'To–ø 100 —Å–ª–æ–≤ –≤ —Ç–æ–∫—Å–∏—á–Ω—ã—Ö –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏—è—Ö')
```

```python
%%time
# –°–æ–∑–¥–∞—ë–º –≤—ã–±–æ—Ä–∫–∏

features_train, features_valid, target_train, target_valid = train_test_split(features, 
                                                                              df['toxic'], 
                                                                              test_size=0.2, 
                                                                              random_state=SEED)
features_valid, features_test, target_valid, target_test = train_test_split(features_valid, 
                                                                            target_valid, 
                                                                            test_size=0.5, 
                                                                            random_state=SEED)
# –í–µ–∫—Ç–æ—Ä–∏–∑–∞—Ü–∏—è —Ç–µ–∫—Å—Ç–∞
vectorizer = TfidfVectorizer(ngram_range=(1,2), max_features=50000)#, stop_words=stopwords)

features_train_vect = vectorizer.fit_transform(features_train)
features_valid_vect = vectorizer.transform(features_valid)
features_test_vect = vectorizer.transform(features_test)

# –ø—Ä–æ–≤–µ—Ä–∏–º —Ä–∞–∑–º–µ—Ä—ã –≤—ã–±–æ—Ä–æ–∫
pd.DataFrame(data=[features_train_vect.shape, target_train.shape,
                   features_valid_vect.shape, target_valid.shape,
                   features_test_vect.shape,  target_test.shape
                  ],
             index=['features_train', 'target_train', 
                    'features_valid', 'target_valid', 
                    'features_test',  'target_test'],
             columns=['–ò–Ω–¥–µ–∫—Å—ã', '–ö–æ–ª–æ–Ω–∫–∏']
            )
```

## –û–±—É—á–µ–Ω–∏–µ

–û–±—É—á–∏–º –Ω–µ—Å–∫–æ–ª—å–∫–æ –º–æ–¥–µ–ª–µ–π —Å –ø–æ–¥–±–æ—Ä–æ–º –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤:
- –õ–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫—É—é —Ä–µ–≥—Ä–µ—Å—Å–∏—é;
- –ö–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ç–æ—Ä –Ω–∞ –æ—Å–Ω–æ–≤–µ —Ä–µ—à–∞—é—â–µ–≥–æ –¥–µ—Ä–µ–≤–∞;
- CatBoostClassifier.

–î–ª—è –∏—Å–∫–ª—é—á–µ–Ω–∏—è –¥—É–±–ª–∏—Ä–æ–≤–∞–Ω–∏—è –∫–æ–¥–∞ —Å–æ–∑–¥–∞–¥–∏–º –¥–≤–µ —Ñ—É–Ω–∫—Ü–∏–∏:
- –§—É–Ω–∫—Ü–∏—è `get_f1_valid` –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –º–æ–¥–µ–ª—å –∏ –≤—ã–±–æ—Ä–∫–∏, –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –∑–Ω–∞—á–µ–Ω–∏–µ –º–µ—Ç—Ä–∏–∫–∏ F1;
- –§—É–Ω–∫—Ü–∏—è `get_f1_cv` –ø—Ä–∏–Ω–∏–∞–µ—Ç –º–æ–¥–µ–ª—å –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –∏ –≤—ã–±–æ—Ä–∫–∏, –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –ª—É—á—à–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –º–æ–¥–µ–ª–∏ –∏ –ª—É—á—à–∏–π F1.

```python
def get_f1_valid(model, X, y, X_test, y_test):
    """ –ü—Ä–∏–Ω–∏–º–∞–µ–º –º–æ–¥–µ–ª—å,
        –≤–æ–∑–≤—Ä–∞—â–∞–µ–º f1 
    """
    model.fit(X, y)
    predictions = model.predict(X_test)
     
    return round(f1_score(y_test, predictions), 3)
```

<div class="alert alert-warning">
    <h2> –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞ <a class="tocSkip"> </h2>

<b>–ù–µ–∫–æ—Ç–æ—Ä—ã–µ –∑–∞–º–µ—á–∞–Ω–∏—è –∏ —Ä–µ–∫–æ–º–µ–Ω–¥–∞—Ü–∏–∏üí°:</b>

–¢—É—Ç –≤—Å–µ –∂–µ –Ω–µ —Å—Ç–æ–∏—Ç –¥–µ–ª–∞—Ç—å —Ç–∞–∫–æ–µ —è–≤–Ω–æ–µ —É–∫–∞–∑–∞–Ω–∏–µ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω—ã—Ö –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ - –ª—É—á—à–µ –ø–µ—Ä–µ–¥–∞–≤–∞—Ç—å –∏—Ö –æ–±—è–∑–∞—Ç–µ–ª—å–Ω–æ. –§—É–Ω–∫—Ü–∏—è –¥–æ–ª–∂–Ω–∞ –±—ã—Ç—å —Å–∞–º–æ–¥–æ—Å—Ç–∞—Ç–æ—á–Ω–∞, –∞ —Å–µ–π—á–∞—Å –æ–Ω–∞ —Å–ª–∏—à–∫–æ–º –≤–ø–ª–µ—Ç–µ–Ω–∞ –≤ —Å–∞–º –∫–æ–¥ –∏ –ø—Ä–∏ –µ—ë –≤—ã–∑–æ–≤–µ –Ω–µ–ø–æ–Ω—è—Ç–Ω–æ, —á—Ç–æ —Ç—ã —Ç—É–¥–∞ —á—Ç–æ-—Ç–æ –ø–µ—Ä–µ–¥–∞–µ—à—å - —É–º–µ–Ω—å—à–∞–µ—Ç —á–∏—Ç–∞–±–µ–ª—å–Ω–æ—Å—Ç—å –∫–æ–¥–∞ –∏ —Ç–∞–∫ –ª–µ–≥—á–µ —Å–¥–µ–ª–∞—Ç—å –æ—à–∏–±–∫—É –≤ –±—É–¥—É—â–µ–º.

<div class="alert alert-block alert-info">
<b>–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Å—Ç—É–¥–µ–Ω—Ç–∞: </b> –û–ö, –ø–æ–ø—Ä–∞–≤–∏–ª.
</div>

```python
def get_f1_cv(model, 
              parametrs, 
              X=vstack((features_train_vect, features_valid_vect)), 
              y=pd.concat([target_train, target_valid])):
    """ –ü—Ä–∏–Ω–∏–º–∞–µ–º –º–æ–¥–µ–ª—å,
        –≤–æ–∑–≤—Ä–∞—â–∞–µ–º f1 
    """
    validation_set_indexes = [-1]*features_train_vect.shape[0] + [0]*features_valid_vect.shape[0]
    ps = PredefinedSplit(test_fold=validation_set_indexes)
    
    rs = RandomizedSearchCV(model, 
                            parametrs, 
                            scoring='f1', 
                            cv=ps,
                            random_state = SEED,
                            n_jobs=-1
                           )
    rs.fit(X, y)
    
    return round(rs.best_score_, 3), rs.best_params_
```

```python
# –°–æ–∑–¥–∞—ë–º –≤—ã–±–æ—Ä–∫—É –∏–∑ –Ω–∞–±–æ—Ä–∞ –¥–∞–Ω–Ω—ã—Ö
df_sample = df.sample(1000).reset_index(drop=True) 
```

```python
# –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º —Ç–µ–∫—Å—Ç –≤ —Ç–µ–Ω–∑–æ—Ä—ã
# —Ç–∞–∫ –∫–∞–∫ –≤ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–Ω–æ–π –º–æ–¥–µ–ª–∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π —Ä–∞–∑–º–µ—Ä —Ç–µ–Ω–∑–æ—Ä–æ–≤ = 512, –ø—Ä–∏–¥—ë—Ç—Å—è –æ–±—Ä–µ–∑–∞—Ç—å –Ω–∞—à–∏ –¥–∞–Ω–Ω—ã–µ –¥–æ —ç—Ç–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è
tokenized = df_sample['text'].apply(lambda x: tokenizer.encode(x, add_special_tokens=True, truncation=True))
# –ó–∞–ø–æ–ª–Ω–∏–º –æ—Å—Ç–∞—Ç–∫–∏ –¥–∞–Ω–Ω—ã—Ö –Ω—É–ª—è–º–∏
padded = np.array([i + [0]*(512 - len(i)) for i in tokenized.values])
# –ó–∞–¥–∞–¥–∏–º –º–∞—Å–∫—É
attention_mask = np.where(padded != 0, 1, 0)
```

```python
# –ü—Ä–æ–≤–µ—Ä–∏–º —Ä–∞–∑–º–µ—Ä
padded.shape
```

```python
# –°–æ–∑–¥–∞—ë–º —ç–º–±–µ–Ω–¥–∏–Ω–≥–∏
batch_size = 100
embeddings = []
for i in notebook.tqdm(range(padded.shape[0] // batch_size)):
        batch = torch.LongTensor(padded[batch_size*i:batch_size*(i+1)])
        attention_mask_batch = torch.LongTensor(attention_mask[batch_size*i:batch_size*(i+1)])
        
        with torch.no_grad():
            batch_embeddings = model(batch, attention_mask=attention_mask_batch)

        embeddings.append(batch_embeddings[0].numpy())
```

```python
# –°–æ–∑–¥–∞—ë–º –≤—ã–±–æ—Ä–∫–∏ –¥–ª—è –º–æ–¥–µ–ª–∏ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏
features_bert = np.concatenate(embeddings)
target_bert = df_sample['toxic']

features_train_bert, features_test_bert, target_train_bert, target_test_bert = train_test_split(features_bert, 
                                                                                                target_bert, 
                                                                                                test_size=0.5, 
                                                                                                random_state=5555)
```
