### Выбор модели

Последовательно произведём обучение моделей классификации со значениями гиперпараметров по умолчанию:
1. Логистическая регрессия
2. Дерево решений
3. Случайный лес

Для каждой модели произведём расчёт метрики f1 и изучим метрику auc-roc с построением кривой ошибок, и выберем модель, показывающую наибольшую f1-меру.

```python
def plt_roc_curve(clf, features, target):
    """Процедура строит roc-кривую и расчитывает метрику auc-roc"""
    # Расчёт вероятности классификации
    probabilities_test = clf.predict_proba(features)
    probabilities_one_test = probabilities_test[:, 1]
    # Получаем данные для построения графика
    fpr, tpr, thresholds = roc_curve(target, probabilities_one_test) 
    # Выводим кривую auc-roc
    plt.figure()
    plt.plot(fpr, tpr, label='Model')
    # ROC-кривая случайной модели
    plt.plot([0, 1], [0, 1], linestyle='--', label='Random')
    plt.xlim([0,1]) 
    plt.ylim([0,1])
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title("ROC-кривая")
    plt.legend(fontsize=12)
    plt.show()
    # Расчёт метрики auc_roc
    print('auc_roc =', roc_auc_score(target, probabilities_one_test))
```

#### Логистическая регрессия

```python
clf = LogisticRegression(random_state=12345) #, class_weight='balanced')
clf.fit(features_train, target_train)
predictions_valid = clf.predict(features_valid)
# Построение кривой ошибок, расчёт auc-roc и f1
plt_roc_curve(clf, features_valid, target_valid)
print('f1      =',f1_score(target_valid, predictions_valid))
```

#### Дерево решений

```python
clf = DecisionTreeClassifier(random_state=12345)
clf.fit(features_train, target_train)
predictions_valid = clf.predict(features_valid)
# Построение кривой ошибок, расчёт auc-roc и f1
plt_roc_curve(clf, features_valid, target_valid)
print('f1      =',f1_score(target_valid, predictions_valid))
```

#### Случайный лес

```python
clf = RandomForestClassifier(random_state=12345)
clf.fit(features_train, target_train)
predictions_valid = clf.predict(features_valid)
# Построение кривой ошибок, расчёт auc-roc и f1
plt_roc_curve(clf, features_valid, target_valid)
print('f1      =',f1_score(target_valid, predictions_valid))
```

## Борьба с дисбалансом

Для исправления дисбаланса в целевом признаке, применим следующие способы:
1. Увеличение выборки
2. Уменьшение выборки
3. Изменение порога классификации

В итоге выберем лучший способ.

### Увеличение выборки

Увеличим количество строк, содержащих "1" в целевом признаке до количества равного количеству строк содержащих "0". Для этого:
1. Разделим обучающую выборку на положительные и отрицательные объекты.
2. Увеличим количество положительных объектов в 4 раза.
3. Создадим новую выборку путём слияния увеличенных положительных значений с отрицательными значениями.
4. Произведём перемешивание данных выборки.

Затем обучим модель на новой выборке и оценим её качество.

```python
def upsample(features, target, repeat):
    """ Функция принемает признаки, целевой признак, количество повторений единиц.
        Возвращает увеличенную выборку"""
    # Деление признаков на 0 и 1
    features_zeros = features[target == 0]
    features_ones  = features[target == 1]
    target_zeros   = target[target == 0]
    target_ones    = target[target == 1]
    # Формирование фичей и таргета с увеличением единиц
    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
    target_upsampled   = pd.concat([target_zeros] + [target_ones] * repeat)
    # Перемешивание
    features_upsampled, target_upsampled = shuffle(features_upsampled, target_upsampled, random_state=12345)
    
    return features_upsampled, target_upsampled

# Создание увеличенной тренировочной выборки
features_upsampled, target_upsampled = upsample(features_train, target_train, 4)
```

```python
# Проверка баланса новой тренировочной выборки
hist_target(target_upsampled)
```

```python
# Обучим модель на увеличенной выборке
upsampled_clf = best_clf
upsampled_clf.fit(features_upsampled, target_upsampled)
predictions_valid = upsampled_clf.predict(features_valid)
f1_score(target_valid, predictions_valid)
```

**Вывод:** На увеличенной выборке, модель обучилась лучше. Метрика F1 увеличилась на 0,05.

### Уменьшение выборки

Уменьшение выборки будем производить аналогично увеличению. Затем обучим модель на новой выборке и оценим её качество.

```python
def downsample(features, target, fraction):
    """ Функция принемает признаки, целевой признак, коэффициент уменьшения повторений нулей.
        Возвращает уменьшенную выборку"""
    # Деление признаков на 0 и 1
    features_zeros = features[target == 0]
    features_ones  = features[target == 1]
    target_zeros   = target[target == 0]
    target_ones    = target[target == 1]
    # Формирование фичей и таргета с уменьшением нулей
    features_downsampled = pd.concat(
        [features_zeros.sample(frac=fraction, random_state=12345)] + [features_ones])
    target_downsampled = pd.concat(
        [target_zeros.sample(frac=fraction, random_state=12345)] + [target_ones])
    # Перемешивание
    features_downsampled, target_downsampled = shuffle(
        features_downsampled, target_downsampled, random_state=12345)
    
    return features_downsampled, target_downsampled
# Создание уменьшенной тренировочной выборки
features_downsampled, target_downsampled = downsample(features_train, target_train, 0.25)
```

```python
hist_target(target_downsampled)
```

```python
# Обучим модель на уменьшенной выборке
downsampled_clf = best_clf
downsampled_clf.fit(features_downsampled, target_downsampled)
predictions_valid = downsampled_clf.predict(features_valid)
f1_score(target_valid, predictions_valid)
```

**Вывод:** На уменьшенной выборке, качество модели увеличилось не удалось.

### Изменение порога классификации

Для определения оптимального порога классификации, произведём расчёт метрики f1 в цикле для порога классификации от 0,3 до 0,8.

```python
# Расчёт вероятности классификации
probabilities_valid = best_clf.predict_proba(features_valid)
probabilities_one_valid = probabilities_valid[:, 1]
max_f1 = 0
threshold_max_f1 = 0
for threshold in np.arange(0.3, 0.8, 0.02):
    predicted_valid = probabilities_one_valid > threshold
    f1 = f1_score(target_valid, predicted_valid)
    print("Порог = {:.2f} | f1 = {:.3f}".format(threshold, f1))
    if f1 > max_f1:
        max_f1 = f1
        threshold_max_f1 = threshold
print("Максимальный f1={:.3f} достигается при пороге = {:.2f}".format(max_f1,threshold_max_f1))
```

**Вывод:** При использовании способа изменения порога классификации можно улучшить метрики f1 до 0,62.
